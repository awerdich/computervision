```python
import numpy as np
import pandas as pd
```

```python
# Generate pixel coordinates
def random_integers(n, interval, seed):
    seed_sequence = np.random.SeedSequence(seed)
    rng = np.random.default_rng(seed)
    integers = rng.integers(low=min(interval), high=max(interval), size=n)
    return integers

seed_x, seed_y = 111, 222
pixel_range = [25, 2048]
n_locations = 50

df = pd.DataFrame({'X': random_integers(n=n_locations, interval=pixel_range, seed=seed_x),
                   'Y': random_integers(n=n_locations, interval=pixel_range, seed=seed_y)})

# Let's calibrate the pixel values
# We calibrate the microscope with a micrometer scale
# The calibration is 700 pixels / 100 um = 7 pixels/um

factor = 7

df = df.assign(X1=df['X']/factor,  
               Y1=df['Y']/factor)

display(df.head(20))
```
